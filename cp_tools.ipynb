```python
def norm_tc(tc):
    pos = 1
    while re.search("0x[0-9a-f]+", tc) != None:
        addr = re.search("0x[0-9a-f]+", tc).group()
        tc = re.sub(addr, "N"+str(pos), tc)
        pos += 1
    return tc
```

```python
for tc in tcs:
    res = ""
    
    tc = norm_tc(tc)
    leaves, nodes, roots  = tc.split("---\n")
    def norm(x):
        return [a for a in x.split("\n") if a!= '']
    leaves = norm(leaves)
    nodes = norm(nodes)
    roots = norm(roots)
    
    templ = """sppf.leaves.reserve([[N]]);\n\n"""
    res += templ.replace("[[N]]", str(len(leaves)))
    
    for line in leaves:
        tag, name = re.search("(\w+): (\w+)", line).groups()
        templ = """sppf.leaves.push_back({"[[TAG]]"});
SPPF_node * [[NAME]] = &sppf.leaves.back();\n"""
        res += templ.replace("[[TAG]]", tag).replace("[[NAME]]", name)
    
    res += "\n"
    
    alts = ""
    for line in nodes:
        tag, name, _alts = re.search("(\w+): (\w+)\t(.+)", line).groups()
        
        templ = """sppf.nodes.push_back({"[[TAG]]"});
SPPF_node * [[NAME]] = &sppf.nodes.back();\n"""
        res += templ\
            .replace("[[TAG]]", tag)\
            .replace("[[NAME]]", name)
        
        templ = "[[NAME]]->alts = [[ALTS]];\n"
        alts += templ\
            .replace("[[NAME]]", name)\
            .replace("[[ALTS]]", _alts.replace(",}", "}") )
        
    
    res += "\n" + alts + "\n"
    
    for line in roots:
        res += "sppf.roots.push_back([[NAME]]);\n"\
            .replace("[[NAME]]", name)
        
        
        

    print("-"*60)
    print(res)
    print("-"*60)
        
```
